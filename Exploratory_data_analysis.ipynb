```python
corrections = {'Lexus': 'Toyota', 'Subaru': 'Toyota'}
df['make'] = df['make'].replace(corrections)
```

```python
# 5. gharama na ina ya gari
plt.figure(figsize=(20, 8))
plt.xticks(rotation=45)
sns.boxplot(x='make', y='price_value', data=df);
```
