```python
# grafu kwa pamoja
```

```python
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, endpoint=True)
x
```

```python
y1 = np.sin(x)
y1
```

```python
y2 = np.cos(x)
y2
```

```python
plt.plot(x, y1, label='Sine')
plt.plot(x, y2, label='Cosine')
plt.legend()
plt.show()
```

```python
# grafu mbili tofauti kwa pamoja
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y1)

plt.subplot(1, 2, 2)
plt.plot(x, y2)

plt.show()
```
