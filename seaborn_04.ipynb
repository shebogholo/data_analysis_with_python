# Seaborn

```python
import seaborn as sns
```

```python
# pip3 install seaborn
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.shape
```

```python
# count plot

# siku gani wanaenda out
sns.countplot('day', data=df);
```

```python
# bar plot
# jinsia gani wanalipa bili kubwa

sns.barplot(x='sex', y='total_bill', data=df);
```

```python
# box plot
sns.boxplot(x='sex', y='total_bill', data=df);
```

```python
# scatter plot
sns.scatterplot(x='tip', y='total_bill', data=df);
```

```python
sns.lmplot(x='tip', y='total_bill', data=df);
```
